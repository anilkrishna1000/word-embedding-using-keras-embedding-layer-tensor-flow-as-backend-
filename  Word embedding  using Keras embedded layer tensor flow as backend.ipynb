### Word Embedding Techniques using Embedding Layer in Keras

```python
### Libraries USed Tensorflow> 2.0  and keras
```

```python
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'This is Anil',
     'I belong to tirupathi',
     'I am quite good in python programming langauge ',
    'I am a good boy',
     'I am a good developer',
     'I am carrying an 2 years of experience ',
     'my hobbies are playing cricket',]
```

```python
sent
```

```python
### Vocabulary size
voc_size=10000
```

#### One Hot Representation

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr) ## observe the printed array indexes....in the above sentences  I word have some index i,e 3299 which is also same for remaining indexes also
```

### Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=10
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
